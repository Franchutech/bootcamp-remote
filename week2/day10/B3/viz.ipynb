## Gráficos con Matplotlib


```python
import matplotlib.pyplot as plt

# Datos básicos
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, color="red", linestyle="--", marker="o")
plt.title("Relación entre X e Y")
plt.xlabel("Valores de X")
plt.ylabel("Valores de Y")
plt.show()
```

### Gráfico de Barras
Comparación de unidades vendidas por producto.


```python
# Gráfico de Barras
productos = ["A", "B", "C", "D", "E"]
ventas = [50, 80, 30, 100, 60]

plt.bar(productos, ventas, color="skyblue")
plt.title("Ventas por producto")
plt.xlabel("Producto")
plt.ylabel("Unidades vendidas")
plt.show()
```

### Histograma
Distribución de notas de un grupo de estudiantes.


```python
# Histograma
notas = [5, 7, 8, 6, 9, 10, 6, 7, 8, 7, 6, 5, 8, 9, 10, 7, 6, 8, 9, 10]

plt.hist(notas, bins=5, color="green", edgecolor="black")
plt.title("Distribución de notas")
plt.xlabel("Rango de notas")
plt.ylabel("Cantidad de estudiantes")
plt.show()
```

### Subplots
Comparación entre la función seno y coseno en la misma figura.


```python
import numpy as np

# Datos
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear figura con 2 gráficos (1 fila, 2 columnas)
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Gráfico seno
axs[0].plot(x, y1, color="blue")
axs[0].set_title("Función Seno")
axs[0].set_xlabel("x")
axs[0].set_ylabel("sin(x)")

# Gráfico coseno
axs[1].plot(x, y2, color="red")
axs[1].set_title("Función Coseno")
axs[1].set_xlabel("x")
axs[1].set_ylabel("cos(x)")

plt.tight_layout()
plt.show()
```

## Gráficos con Seaborn


```python
import seaborn as sns

# Dataset de ejemplo
tips = sns.load_dataset("tips")

# Gráfico categórico: total de la cuenta por día
sns.catplot(data=tips, x="day", y="total_bill", kind="strip", jitter=True)
plt.show()
```

### Distribución de propinas
Histograma de la variable "tip" con curva de densidad.


```python
# Distribución de propinas
sns.histplot(data=tips, x="tip", kde=True, color="orange")
plt.title("Distribución de propinas")
plt.xlabel("Monto de la propina")
plt.ylabel("Frecuencia")
plt.show()
```

### Heatmap de correlaciones
Relaciones entre las variables numéricas del dataset `tips`.


```python
# Matriz de correlación del dataset tips
corr = tips.corr(numeric_only=True)

# Heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Mapa de calor de correlaciones")
plt.show()
```
